```python
def jos(n, k):
	l = n - max(k**i for i in range(100) if k**i <= n)
	return k*l, l
jos(14,3)
```

```python
from numpy import prod
def fac(n):
    x = [i for i in range(1,n+1) if not n%i]
    y = [x[i] for i in range(1,len(x)) if int(x[i]**.5)**2 == x[i]]
    return int(prod(y)**.5), n//prod(y)
fac(72)
```

```python
def polybius(txt, lets='abcdefghiklmnopqrstuvwxyz'):
    poly = [list(lets[:5]),list(lets[5:10]),list(lets[10:15]),list(lets[15:20]),list(lets[20:25]),[' ']]
    txt = txt.lower()
    if txt[0] not in lets:
        txt = txt.replace(' ','60')
        txt = [txt[i:i+2] for i in range(0,len(txt),2)]
        print(''.join([poly[int(el[0])-1][int(el[1])-1] for el in txt]))
    else:
        txt.replace('j','i')
        h = []
        for c in txt:
            for i in range(6):
                try:
                    h.append(str(i+1)+str(1+poly[i].index(c)))
                    break
                except:
                    pass
        print(''.join(h).replace('61',' '))
```

```python
from numpy import prod
def odd(*args):
    x = list(str(sum(args)))
    return prod([int(u) for u in x]) if prod([int(u) for u in x]) < 10 else odd(prod([int(u) for u in x]))

odd(1,3,4,53,2,3,5,3,7)
```

```python
class Food:
    def __init__(self,n):
        self.f = n
        self.fs = str(n)
        self.b = int(self.fs[::-1])
        self.bs = str(self.b)
        self.fdish = [int(str(self.f)[i:]) for i in range(len(str(self.f)))]
        self.bdish = [int(str(self.b)[i:]) for i in range(len(str(self.b)))]
        self.ftaste = sum(self.fdish)
        self.btaste = sum(self.bdish)
        
Food(2134).bdish
Food(2134).fdish
```

```python
def next_prime(n):
    for i in range(2,int(n**.5)+1):
        if n%i:
            return n
        else:
            return 4

next_prime(53)
```

```python
# a new lead on the primes
def pchk(n):
    if not n%2:
        return [x for x in range(((n-1)**2)+2,(n**2)-1,2) if all([x%i for i in range(2,n)])] 
    else:
        return [x for x in range(((n-1)**2)+1,(n**2)-1,2) if all([x%i for i in range(2,n)])]
```

```python
a = timer()
for i in range(101):
    print(pchk(i))
b = timer()
print(b-a)
```

```python
def mats(n,i=0,m=[]):
    while i != n:
        m.append([x*0 for x in range(n)])
        i += 1
    for j in range(n):
        m[j][j] = 1
    return m
mats(7)
```

```python
import numpy as np
def convolve(sequence, weights):
    convolution = np.zeros(len(sequence) - len(weights) + 1)
    for i in range(convolution.shape[0]):
        convolution[i] = np.sum(np.dot(weights,sequence[i:i+len(weights)]))
        
s = [2,3,5,7,11]
w = [-1,1]
print(convolve(s,w))
```

```python
def vowel_links(txt,vtv=False,v='aeiou'):
	vtv = True if txt.split()[x][-1] in v and txt.split()[x+1][0] in v for x in range(len(txt.split())-1)		
	return vtv

vowel_links('the sudden applause')
```
